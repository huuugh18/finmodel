```python
import pandas as pd
temp_data = [
    ['Monday', 10, 80, 15,  1],
    ['Tuesday', 10, 90, 15,  2 ],
    ['Wednesday', 5, 40, 30,  0],
    ['Thursday', 2, 30, 25,  0],
    ['Friday', 12, 90, 25,  5]
]

df = pd.DataFrame(temp_data, columns=['Day Of Week', 'Temperature (degrees)', 'Humiditiy(%)', 'Wind Speed (mph)', 'Rainfall (cm)'])
df.set_index('Day Of Week')
```

### Turn the following data into a data frame and plot the data (line graphs, scatter plots) so that the graph provides useful visual information. 

```python
df.plot()
```

### Plot One Column

```python
df.iloc[:,2].plot()
```

```python
df.plot.scatter(x='Humiditiy(%)', y='Wind Speed (mph)')
```
