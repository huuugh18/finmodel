```python
import yfinance as yf
```

```python
snow_frame = yf.Ticker('SNOW').history(start='2021-01-01', end='2021-12-01')
snow_prices = snow_frame['Close'].to_frame()
snow_prices.plot()
```

Can see that this data is trending - not stationary.
Will take the derivative. This is a discrete view of a derivative. It's what economistst refer to as marginal values.  
Going to have the *marginal prices* instead of the prices themselves.  
**Marginal Prices** are the difference b/w yesterdays price and todays price.

Need to create a new column in the dataframe.  
Use the `.shift()` command to get the lagged value 

```python
snow_prices['Differences'] = snow_prices['Close'] - snow_prices['Close'].shift()
```

```python
snow_prices.plot()
```

Can see that the differences is stationary. 

Use this differences series to do our predictions then convert predictions back to regular series.
